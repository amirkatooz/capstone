```python
import pandas as pd
import numpy as np
```

```python
import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import zipfile
```

```python
mixed_data_final = pd.read_pickle('mixed_data_final.data')
```

```python
X = mixed_data_final['Image'].apply(lambda image: np.fromstring(image, sep=' '))
y = mixed_data_final.drop('Image', axis=1).values
```

```python
X.shape
```

```python
X = np.vstack(X) / 255
y = (y - 48) / 48
```

```python
from keras.layers import Input, Conv2D, MaxPool2D, Dropout, Activation, Dense, Flatten
from keras.models import Sequential
```

```python
cnn = Sequential([
    Conv2D(32, kernel_size=(3, 3),
           activation='relu',
           input_shape=(96,96,1)),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(.2),
    
    Conv2D(64, kernel_size=(2, 2),
           activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    Dropout(.3),  
    
    Flatten(),
    Dense(1000, activation='relu'),

    Dense(8)
])
```

```python
cnn.compile(loss='mse', optimizer='sgd')
```

```python
X2 = X.reshape(-1, 96, 96, 1)
```

```python
cnn.fit(X2, y, validation_split=.2, epochs=200)
```

```python
cnn.save('cnn4.h5')
```

```python
cnn.evaluate(X2, y)
```

```python
from keras.models import load_model
cnn1 = load_model('cnn1.h5')
```

```python
cnn1.evaluate(X2, y)
```
