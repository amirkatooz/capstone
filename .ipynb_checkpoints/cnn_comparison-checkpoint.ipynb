```python
import pandas as pd
import numpy as np
import zipfile
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import keras
```

```python
mixed_data_final = pd.read_pickle('mixed_data_final.data')
```

```python
X = mixed_data_final['Image'].apply(lambda image: np.fromstring(image, sep=' '))
y = mixed_data_final.drop('Image', axis=1).values
```

```python
X = (np.vstack(X) / 255).reshape(-1, 96, 96, 1)
y = (y - 48) / 48
```

```python
from keras.layers import Input, Conv2D, MaxPool2D, Dropout, Activation, Dense, Flatten
from keras.models import Sequential
from keras import optimizers
```

# CNN 5

```python
cnn5 = Sequential([
    Conv2D(40, kernel_size=(4, 4),
           activation='tanh',
           input_shape=(96,96,1)),
    MaxPool2D(pool_size=(2, 2)),
    
    Conv2D(80, kernel_size=(3, 3),
           activation='tanh'),
    MaxPool2D(pool_size=(2, 2)),

    Conv2D(120, kernel_size=(3, 3),
           activation='tanh'),
    MaxPool2D(pool_size=(2, 2)),
    
    Flatten(),
    Dense(120, activation='tanh'),

    Dense(8)
])
```

```python
n_epochs=200

cnn5.compile(loss='rmse', optimizer='sgd')
```

```python
history5 = cnn5.fit(X, y, validation_split=.2, epochs=n_epochs)
```

```python
cnn5.save('cnn5.h5')
```

```python
cnn5_rmse = cnn5.evaluate(X, y) * 48
cnn5_rmse
```

# CNN 6

```python
cnn6 = Sequential([
    Conv2D(40, kernel_size=(4, 4),
           activation='tanh',
           input_shape=(96,96,1)),
    MaxPool2D(pool_size=(2, 2)),
    
    Conv2D(80, kernel_size=(3, 3),
           activation='tanh'),
    MaxPool2D(pool_size=(2, 2)),

    Conv2D(120, kernel_size=(3, 3),
           activation='tanh'),
    MaxPool2D(pool_size=(2, 2)),
    
    Flatten(),
    Dense(120, activation='tanh'),

    Dense(8)
])
```

```python
n_epochs=200
start_learning_rate = .03
stop_learning_rate = .001
decay_rate = (start_learning_rate - stop_learning_rate) / n_epochs

sgd = optimizers.SGD(lr=learning_rate, decay=decay_rate)

cnn6.compile(loss='rmse', optimizer=sgd)
```

```python
history6 = cnn6.fit(X, y, validation_split=.2, epochs=n_epochs)
```

```python
cnn6.save('cnn6.h5')
```

```python
cnn6_rmse = cnn6.evaluate(X, y) * 48
cnn6_rmse
```
