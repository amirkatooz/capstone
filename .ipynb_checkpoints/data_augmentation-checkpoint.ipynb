```python
import pandas as pd
import numpy as np
import zipfile
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from lib.image_generator import adjust_contrast, flip_images, rotate_images
```

```python
with zipfile.ZipFile('data/training.zip') as zip:
    with zip.open('training.csv') as myZip:
        df = pd.read_csv(myZip)
```

```python
df_head = df.head()
```

```python
rotated_cw_5 = rotate_images(df_head, 20)
```

```python
f, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(np.fromstring(df.loc[0, 'Image'], sep=' ').reshape(96, 96), cmap='gray')
ax[1].imshow(np.fromstring(rotated_cw_5.loc[0, 'Image'], sep=' ').reshape(96, 96), cmap='gray')
```

```python
flipped_df = flip_images(df_head)
```

```python
f, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(np.fromstring(df.loc[1, 'Image'], sep=' ').reshape(96, 96), cmap='gray')
ax[1].imshow(np.fromstring(flipped_df.loc[1, 'Image'], sep=' ').reshape(96, 96), cmap='gray')
```

```python
adjusted_contrast_df = adjust_contrast(df_head, degree=.5)
```

```python
f, ax = plt.subplots(1, 2, figsize=(8, 8))
ax[0].imshow(np.fromstring(df.loc[2, 'Image'], sep=' ').reshape(96, 96), cmap='gray', vmin=0, vmax=255)
ax[1].imshow(np.fromstring(adjusted_contrast_df.loc[2, 'Image'], sep=' ').reshape(96, 96), cmap='gray', vmin=0, vmax=255)
```
